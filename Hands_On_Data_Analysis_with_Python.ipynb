## Eksplorasi data : Agregasi 

###Grouping satu kolom

Perintah grouping dengan `groupby`, menghasilkan object `DataFrameGroupBy` yang belum akan dieksekusi sampai ada operasi yang menampilkan hasilnya.

```python
df.groupby(['hotel'])
```

Misalnya menampilkan `count`. Perintah `count()` pada pandas menghitung jumlah baris dari seluruh kolom yang ada

```python
df.groupby(['hotel']).count()
```

Untuk menampilkan jumlah baris saja, bisa digunakan fungsi `size()`

```python
df.groupby(['hotel']).size()
```

###Grouping dengan multikolom

Untuk group by terhadap lebih dari 1 kolom, tulis nama kolom dalam list parameter.
Hasilnya berupa dataframe dengan `multi index`

```python
df.groupby(['hotel','reservation_status']).size()
```

Menghindari multiindexing, dengan set option `as_index = False`

```python
df.groupby(['hotel','reservation_status'], as_index=False).size()
```

Menghindari multiindexing bisa juga dilakukan dengan fungsi `reset_index()`

```python
df.groupby(['hotel','reservation_status']).size().reset_index()
```

###Agregasi lainnya

Simpan hasil grouping ke dalam variabel agar lebih efektif

```python
grouped = df.groupby(['hotel','reservation_status'])
```

Multiple agregasi terhadap sebuah kolom, menggunakan fungsi `agg()`
Parameternya bisa berupa fungsi, dalam hal ini kita gunakan fungsi-fungsi `numpy` 

```python
grouped['stays_in_weekend_nights'].agg([np.sum, np.mean, np.min, np.max])
```

Bisa juga menyebutkan nama fungsinya

```python
grouped['stays_in_weekend_nights'].agg(['sum', 'mean', 'min', 'max'])
```

Untuk agregasi terhadap multipel kolom, masukkan nama-nama kolom yang akan diagregasi ke dalam list.

Hasilnya berupa multi-indeks baris dan kolom. 

```python
grouped[['stays_in_weekend_nights','stays_in_week_nights']].agg([np.sum, np.mean, np.min, np.max])
```

Untuk menghindari duplikasi nama kolom, kita bisa memberi nama kolom hasil agregasi

```python
grouped.agg(
     max_weekend=('stays_in_weekend_nights', 'max'),
     avg_weekend=('stays_in_weekend_nights', 'mean'),
     total_weekend=('stays_in_weekend_nights', np.sum),
     max_week=('stays_in_week_nights', 'max'),
     avg_week=('stays_in_week_nights', 'mean'),
     total_week=('stays_in_week_nights', 'sum')
)
```

##Enrichment : Join dengan referensi

Load data negara ke dalam pandas dataframe

```python
df_ref = pd.read_csv('https://raw.githubusercontent.com/lukes/ISO-3166-Countries-with-Regional-Codes/master/all/all.csv')
df_ref
```

Join dengan data reservasi, kita pilih kolom `name`,`region`, dan `sub-region` dari data referensi.

```python
df_merged = pd.merge(df, df_ref[['alpha-3','name','region','sub-region']], 
                     left_on='country', 
                     right_on = 'alpha-3', 
                     how='left')
```

Menangani NULL 

```python
 df_merged = df_merged.fillna({'name':'UNK', 'region':'UNK', 'sub-region':'UNK'})
```

##Visualisasi

Kita bisa menggunakan fungsi `plot()` yang terdapat pada pandas DataFrame

```python
df_merged.groupby(['hotel']).agg(
    total_stays=('stays_in_total', np.sum)).plot.bar()
```

Atau menggunakan `seaborn`

```python
sns.barplot(data=df_merged, x="hotel", y="stays_in_total", ci=None)
```

Plot jumlah reservasi berdasar jenis hotel dan status 

```python
plt.figure(figsize=(8, 6))

sns.countplot(x="reservation_status",
              hue="hotel",
              data=df_merged
)
plt.title('Total Reservation by Status')
plt.show()
```

Plot rata-rata lama menginap berdasar jenis hotel dan status 

```python
dfm = df_merged.groupby(['hotel','reservation_status']) \
      .agg(avg_all=('stays_in_total', 'mean')) \
      .reset_index()

plt.figure(figsize=(8, 6))

sns.barplot(x="reservation_status", 
            y='avg_all', 
            hue="hotel", 
            data=dfm, 
            order=['Check-Out','Canceled','No-Show'])

plt.title('Average Stays-in by Status')
plt.show()
```

Kita bisa menampilkan beberapa plot dalam satu grafis dengan menggunakan `plt.subplot()`

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

sns.countplot(x="reservation_status",
              hue="hotel",
              data=df_merged,
              order=['Canceled','Check-Out','No-Show'],
              hue_order=['City Hotel','Resort Hotel'],
              ax=axes[0]
)
axes[0].set_title('Total Reservation by Status')

dfm = df_merged.groupby(['hotel','reservation_status']) \
      .agg(avg_stay=('stays_in_total', 'mean'), 
           total_stay=('stays_in_total', 'sum')) \
      .reset_index()

sns.barplot(x="reservation_status",
            y="total_stay",
            hue="hotel",
            data=dfm,
            ax=axes[1]
)
axes[1].set_title('Total Stays-in by Status')

sns.barplot(x="reservation_status", 
            y='avg_stay', 
            hue="hotel", 
            data=dfm,             
            ax=axes[2]
)
axes[2].set_title('Average Stays-in by Status')

plt.show()
```
